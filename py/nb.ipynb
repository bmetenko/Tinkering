```python
import panel as pn
pn.extension()
pn.extension('echarts')
from pyecharts import options as opts
from pyecharts.charts import Bar
a = (
    Bar()
        .add_xaxis(["a", "b", "c", "d", "e", "f"])
        .add_yaxis("aa", [5, 20, 36, 10, 75, 90])
        .add_yaxis("bb", [15, 25, 16, 55, 48, 8])
        .set_global_opts(title_opts=opts.TitleOpts(title="Bar-aaa", subtitle="bbb"))
    )
echart_pane = pn.pane.ECharts(echart, height=480, width=640)
echart_pane
```

```python
import numpy as np, pandas as pd, matplotlib.pyplot as plt
%matplotlib inline

def mplplot(df, **kwargs):
    fig = df.plot().get_figure()
    plt.close(fig)
    return fig

def sine(frequency=1.0, amplitude=1.0, n=200, view_fn=mplplot):
    xs = np.arange(n)/n*20.0
    ys = amplitude*np.sin(frequency*xs)
    df = pd.DataFrame(dict(y=ys), index=xs)
    return view_fn(df, frequency=frequency, amplitude=amplitude, n=n)

sine(1.5, 2.5)
```
